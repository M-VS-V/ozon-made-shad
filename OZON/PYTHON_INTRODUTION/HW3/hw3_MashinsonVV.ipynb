```python
from math import pi, inf
import random

def double_equal(left, right, eps = 1e-6):
    return abs(left - right) < eps

def distance(coords1, coords2):
    return ((coords1[0] - coords2[0]) ** 2 + (coords1[1] - coords2[1]) ** 2) ** 0.5
```

Треугольник
    Написать класс для треугольника на плоскости Triangle. В классе всего 3 поля - длины
    сторон. Нужно написать следующие методы:
    1. Конструктор. (__init__(self, l1, l2, l3)). Должен проверять, можно ли создать такой
    треугольник. Если нельзя - бросает эксепшн (материал об этом будет на будущей
    лекции).
    2. Периметр (perimeter(self)). Считает и возвращает периметр треугольника.
    3. Площадь (area(self)). Считает и возвращает площадь.
    4. Равнобедренный ли (isosceles(self)). Возвращает True, если треугольник
    равнобедренный.
    5. Равносторонний ли (equilateral(self)). Возвращает True, если треугольник
    равносторонний.
    6. Равен ли другому треугольнику. (Перегрузить проверку на равенство)

```python
class TriangleNotExistException(Exception):
    pass

class Triangle:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
        ordered = list(sorted([a, b, c]))
        if ordered[0] + ordered[1] <= ordered[2]:
            raise TriangleNotExistException
            
    def ordered_sides(self):
        return list(sorted([self.a, self.b, self.c]))
        
    def __repr__(self):
        return f"Triangle: {self.a} {self.b} {self.c}"
    
    def __eq__(self, other):
        return self.ordered_sides() == other.ordered_sides()
    
    def __ne__(self, other):
        return not (self == other)
    
    
    def perimeter(self):
        return self.a + self.b + self.c
        
    
    def area(self):
        pp = self.perimeter() / 2.0
        return (pp * (pp - self.a) * (pp - self.b) * (pp - self.c)) ** 0.5
        
    
    #Равнобедренность
    def isosceles(self):
        return self.a == self.b or self.b == self.c or self.a == self.c
        
    #Равносторонность
    def equilateral(self):
        return self.a == self.b == self.c
        
        
        
assert(Triangle(1, 3, 3) == Triangle(3, 3, 1))
assert(Triangle(1, 3, 3) != Triangle(3, 3.5, 1))
assert(double_equal(Triangle(3, 4, 5).area(), 6.0))
assert(double_equal(Triangle(3, 4, 5).perimeter(), 12.0))
assert(Triangle(3, 3, 1).isosceles())
assert(not Triangle(3, 4, 2).isosceles())
assert(Triangle(3, 3, 3).equilateral())
assert(not Triangle(3, 3, 2).equilateral())
```

Окружность
Написать класс для окружности на плоскости Circle. У класса два поля - координаты и
радиус. Координаты задаются массивом из двух чисел. Нужно написать методы:
1. Конструктор со значениями по умолчанию (__init__(self, coords, r))
2. Площадь (area(self)). Считает и возвращает площадь.
3. Периметр (perimeter(self)). Считает и возвращает периметр.
4. Переместить круг в конкретную точку (move_to_point(self, coords))
5. Изменить радиус (set_radius(self, r))
6. Определить, пересекается ли окружность с другой (перегрузить &)
7. Определить, находится окружность внутри другой полностью (перегрузить <, >, =)

```python
class Circle:
    def __init__(self, coords = [0, 0], r = 1):
        self.coords = coords
        self.r = r
    
    
    def perimeter(self):
        return 2 * pi * self.r
        
    
    def area(self):
        return pi * self.r ** 2
    
    
    def move_to_point(self, coords):
        self.coords = coords
        
        
    def set_radius(self, r):
        self.r = r
        
        
    def __and__(self, other):
        r1 = max(self.r, other.r)
        r2 = min(self.r, other.r)
        center_distance = distance(self.coords, other.coords)
        return center_distance <= r1 + r2 and center_distance >= r1 - r2
    
    
    def __lt__(self, other):
        return self.r < other.r and distance(self.coords, other.coords) < other.r - self.r
    
    
    def __gt__(self, other):
        return other < self
    
    
    def __eq__(self, other):
        return self.r == other.r and self.coords == other.coords
    
    
    def __ne__(self, other):
        return not (self == other)

        
    
assert(double_equal(Circle(r = 2).area(), 4 * pi))
assert(double_equal(Circle(r = 2).perimeter(), 4 * pi))
c = Circle(r = 2)
c.set_radius(3)
c.move_to_point([1, 2])
assert(c.r == 3 and c.coords == [1, 2])
assert(Circle() == Circle())
assert(Circle(coords = [1, 2], r = 3) == Circle(coords = [1, 2], r = 3))
assert(Circle(coords = [1, 2], r = 4) != Circle(coords = [1, 2], r = 3))
assert(Circle() < Circle(r = 10))
assert(Circle(r = 10) > Circle())
assert(Circle(r = 10) & Circle(coords = [5, 0], r = 5))
assert(Circle(r = 10) & Circle(coords = [15, 0], r = 5))
assert(Circle(r = 10) > Circle(coords = [4, 0], r = 5))
assert(not(Circle(r = 10) & Circle(coords = [16, 0], r = 5)))
```

Товар
Написать класс Item для различных товаров, которые могут быть в магазине (нужно для
следующего задания). Например компьютер, экран, мышка, флешка, телефон. Атрибуты
передаются в конструкторе.

```python
class Item:
    def __init__(self, name, price, category):
        self.name = name
        self.price = price
        self.category = category
        
    def __eq__(self, other):
        return self.name == other.name and self.category == other.category
    
    def __ne__(self, other):
        return not (self == other)
        
        
    def __repr__(self):
        return f"Name: {self.name}, price: {self.price}, category = {self.category}"
    
        
class Shop:
    def __init__(self):
        self.items = []
        
        
    def add(self, item):
        self.items.append(item)
        
        
    def filtered_by_categories(self, categories):
        return list(filter(lambda x: categories is None or x.category in categories, self.items))
        
        
    def count_category(self, category):
        return len(self.filtered_by_categories([category]))
    
    
    def find_items(self, price_bounds, categories: list):
        lb = price_bounds[0] if price_bounds[0] is not None else -inf
        rb = price_bounds[1] if price_bounds[1] is not None else inf
        
        def is_satisfy_condition(item: Item):
            is_good_category = (categories is None) or (item.category in categories)
            is_good_price = item.price >= lb and item.price <= rb
            return is_good_category and is_good_price
        
        return list(filter(lambda x: is_satisfy_condition(x), self.items))
    
    def get_full_price(self, categories):
        items = self.filtered_by_categories(categories)
        return sum([item.price for item in items])
    
    
    def is_available(self, item):
        return item in self.items
    
shop = Shop()
shop.add(Item("name_1", 1, "cat_1"))
shop.add(Item("name_2", 2, "cat_1"))
shop.add(Item("name_3", 3, "cat_2"))
shop.add(Item("name_4", 100, "cat_4"))
assert(shop.count_category("cat_1") == 2)
assert(shop.count_category("cat_2") == 1)
assert(shop.count_category("cat_3") == 0)
assert(shop.get_full_price(None) == 106)
assert(shop.get_full_price(["cat_1"]) == 3)
assert(shop.is_available(Item("name_1", 1, "cat_1")))
```

Написать класс робота Robot, который может драться с другим роботом. За это будет
отвечать метод robot1.fightWith(robot2).

```python
def robot_hit_decorator(function):
    def wrapper_accepting_arguments(robot1, robot2):
        damage = function(robot1, robot2)
        resultStr = f"Робот {robot1.name} ударил робота {robot2.name} и нанес {damage} урона здоровью. "
        resultStr += f"Здоровье робота {robot1.name}: {robot1.health} "
        resultStr += f"здоровье робота {robot2.name}: {robot2.health}"
        print(resultStr)
    return wrapper_accepting_arguments


class Robot:
    count = 1
    
    def __init__(self, name=None, power=10):
        if name is None:
            self.name = f"Robot {Robot.count}"
        else:
            self.name = name
            
        Robot.count +=1
        
        self.power = power
        self.health = 100
        
    @robot_hit_decorator
    def hit(self, robot):
        damage = random.randint(1, self.power)
        robot.health = max(0, robot.health - damage)
        return damage

    def fightWith(self, robot):
        hit_order = 0
        while self.health > 0 and robot.health > 0:
            if hit_order % 2 == 0:
                attacker = self
                attacked = robot
                damage = self.hit(robot)
            else:
                attacker = robot
                attacked = self
                damage = robot.hit(self)
            
            hit_order += 1
            hit_order %= 2
        
        if self.health == 0:
            print(f"Победил {robot.name}")
        else:
            print(f"Победил {self.name}")
            
        
r1 = Robot()
r2 = Robot()

r1.fightWith(r2)
```
